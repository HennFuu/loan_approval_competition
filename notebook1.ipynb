```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./train.csv')
df.head(5)
```

```python
df.select_dtypes(include='object')
```

```python
for cat in df.select_dtypes(include='object').columns.tolist():
    plt.title(cat)
    sns.countplot(data=df, x=cat,hue='loan_status')
    plt.show()
```

```python
cat_df = df.select_dtypes(include=['int','float']).drop(columns=['id'])
cat_df.head(5)
```

```python
sns.pairplot(cat_df)
plt.show()
```

```python
sns.pairplot(cat_df, hue='loan_status')
plt.show()
```
